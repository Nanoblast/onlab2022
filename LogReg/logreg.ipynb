```python
x = np.vstack((
    np.vstack((np.random.randn(100)-1,np.random.randn(100)-2)).T,
    np.vstack((np.random.randn(100),np.random.randn(100))).T
))
y = np.concatenate((np.zeros(100),np.ones(100)))
plt.scatter(x[:,0],x[:,1],c=y);
```

```python
def sigmoid(a):
  return 1/(1+np.exp(-a))

def sigmoid(a):
  u = np.exp(-np.abs(a))
  return np.where(a>0,1/(1+u),u/(1+u))

w = np.random.randn(7)
ùöΩ = np.hstack([np.ones((len(x),1)),x,x**2,x**3])

gr     = np.linspace(-4,3,100)
gx,gy  = np.meshgrid(gr,gr)
xt = np.vstack([gx.flatten(),gy.flatten()]).T
ùöΩt = np.hstack([np.ones((len(xt),1)),xt,xt**2,xt**3])

for i in range(1000):
  d = y - sigmoid(ùöΩ @ w)
  g = - ùöΩ.T @ d
  w -= 0.01*g

pred = sigmoid(ùöΩt @ w)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.contour(gx,gy,pred.reshape(100,100),levels=[0.5])
```
